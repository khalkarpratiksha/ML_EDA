```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Set the style
sns.set_style("darkgrid")

# Example: Bar plot of attack types vs. their counts
plt.figure(figsize=(5, 6))
sns.barplot(x=df["Attack Type"].value_counts().index, 
            y=df["Attack Type"].value_counts().values, 
            palette="viridis")

# Labels and title
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.title("Frequency of Different Cybersecurity Attacks")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

```python

# Set the style for the plot
sns.set_style("darkgrid")

# Create the scatter plot (Example: Anomaly Scores vs. Packet Length)
plt.figure(figsize=(6, 6))
sns.scatterplot(x=df["Anomaly Scores"], y=df["Packet Length"], hue=df["Severity Level"], palette="coolwarm", size=df["Severity Level"], sizes=(20, 200))

# Adding labels and title
plt.xlabel("Anomaly Score", fontsize=12)
plt.ylabel("Packet Length", fontsize=12)
plt.title("Scatter Plot of Anomaly Scores vs. Packet Length", fontsize=14)

# Show the plot
plt.show()
```
